```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
print('Esperamos trabalhar no diretório')
print(os.getcwd())
```

```python
#Se usar o arquivo descompactado
dados=pd.read_csv('DOM2015.csv',sep=',')
dados.shape
#dados
#Se usar o arquivo compactado como está no dropbox púbico
#pd.read_csv('DOM2015csv.zip', compression='zip', header=0, sep=',', quotechar='"')
```

```python
df=pd.DataFrame({'Tipo_do_Domicílio':dados.V0202})
df['TV_em_cores']=dados.V0226
df['TV_por_assinatua']=dados.V02272
df['Tablet']=dados.V02321
df['Acesso_a_internet_em_seu_domiciílio']=dados.V02322
df['Renda_per_capita']=dados.V4621
df.Tipo_do_Domicílio=df.Tipo_do_Domicílio.astype('category')
df.Tipo_do_Domicílio.cat.categories=('Casa','Apartamento','Cômodo')
df.TV_em_cores=df.TV_em_cores.astype('category')
df.TV_em_cores.cat.categories=('Sim','Não')
df.TV_por_assinatua=df.TV_por_assinatua.astype('category')
df.TV_por_assinatua.cat.categories=('Sim','Não')
df.Tablet=df.Tablet.astype('category')
df.Tablet.cat.categories=('Sim','Não')
df.Acesso_a_internet_em_seu_domiciílio=df.Acesso_a_internet_em_seu_domiciílio.astype('category')
df.Acesso_a_internet_em_seu_domiciílio.cat.categories=('Sim','Não')
df.dropna()
```
