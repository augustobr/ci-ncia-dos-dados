### Pretendo analisar o aumento, ou diminuição, das TVs por assinatura e de internet no Brasil ao longo dos anos (2013-2015), também pretendo analisar a preferência de um brasileiro que não possui uma renda alta, se ele prefere ter TV por assinatura ou internet em seu domicílio.

## Análise e Resultados

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
print('Esperamos trabalhar no diretório')
print(os.getcwd())
```

```python
#Se usar o arquivo descompactado
dados=pd.read_csv('DOM2015.csv',sep=',')

df=pd.DataFrame({'TV_por_assinatua':dados.V02272})

df['TV_por_assinatua']=dados.V02272

df['Acesso_a_internet_em_seu_domiciílio']=dados.V02322
df['Renda_per_capita']=dados.V4621

df.TV_por_assinatua=df.TV_por_assinatua.astype('category')
df.TV_por_assinatua.cat.categories=('Sim','Não')

df.Acesso_a_internet_em_seu_domiciílio=df.Acesso_a_internet_em_seu_domiciílio.astype('category')
df.Acesso_a_internet_em_seu_domiciílio.cat.categories=('Sim','Não')
df.Renda_per_capita=df.Renda_per_capita.astype('float')
df=df[(df.Renda_per_capita<4000)].dropna()
```

```python
plt.subplot(121)
ut1 = df.TV_por_assinatua.value_counts().reindex(['Sim', 'Não'])
ut1 = (df.TV_por_assinatua.value_counts(normalize=True)*100).round(decimals=1).reindex(['Sim', 'Não'])
plot = ut1.plot(kind='pie',autopct="%.2f",title='TV por assinatura 2015',figsize=(15,5))
plt.ylabel('Freq.Relativa (em%)')
plt.xlabel('Possui TV por assinatura')

plt.subplot(122)

ut2 = df.Acesso_a_internet_em_seu_domiciílio.value_counts().reindex(['Sim', 'Não'])
ut2 = (df.Acesso_a_internet_em_seu_domiciílio.value_counts(normalize=True)*100).round(decimals=1).reindex(['Sim', 'Não'])

plot = ut2.plot(kind='pie',autopct="%.2f",title='Acesso a internet em seu domiciílio 2015',figsize=(15,5))
plt.ylabel('Freq.Relativa (em%)')
plt.xlabel('Possui acesso a internet')
```

## Conclusão 

### Pode-se concluir que, mesmo sendo pouco, houve um aumento no número de pessoas que possuem TV por assinatura em seu domicílio, porém o aumento no número de pessoas com acesso a intenet é um aumento bem considerável, uma vez que de 46.3% em 2013, foi para 56.9% em 2015. 
### Estes aumentos fazem parecer que a sociedade brasileira está ficando cada vez com mais acesso a internet e TV por assinatura, porém ao montar os histogramas podemos afirmar que apenas as pessoas com uma renda "alta" fizeram esses números aumentarem. As pessoas com baixa renda, como mostrado nos histogramas acima, não aumentaram seu acesso a intenet nem a TV por assinatura, pelo contrário, o número de pessoas de baixa renda sem acesso a internet e sem TV por assinatura aumentou!! 

### Minha conclusão é que, apesar dos números referentes ao acesso a internet e TVs por assinatura terem aumentado, as pessoas de baixa renda estão cada vez mais sem internet em suas casas e muito menos sem TVs por assinatura, e com isso é possível perceber certa desigualdade no Brasil.
